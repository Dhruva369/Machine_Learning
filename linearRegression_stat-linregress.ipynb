```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Advertising and Sales using stat.lineregress

x=[1,2,3,4,5]
y=[1,1,2,2,4]

plt.scatter(x,y)
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc,x))

print(f"Slope = {slope} \nIntercept = {intercept} \nmymodel : {mymodel}")
```

```python
plt.scatter(x,y)
fig=plt.plot(x,mymodel,lw=3,c='purple',linestyle='-')
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

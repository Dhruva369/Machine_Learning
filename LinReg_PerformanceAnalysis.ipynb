```python
import matplotlib.pyplot as plt
from scipy import stats

#Years after purchase and Resale value using stat.lineregress

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x,y)
plt.xlabel("Years after purchase")
plt.ylabel("Resale value")
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc,x))

print(f"Slope = {slope} \nIntercept = {intercept} \nr = {r} \nR-square = {r**2} \nStandard Error = {std_err} \nmymodel : {mymodel}")
```

```python
plt.scatter(x,y)
fig=plt.plot(x,mymodel,lw=3,c='purple')
plt.xlabel("Years after purchase")
plt.ylabel("Resale value")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
#Advertising and Sales using OLS model
x=[1,2,3,4,5]
y=[1,1,2,2,4]

plt.scatter(x,y)
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
x1=sm.add_constant(x)
results=sm.OLS(y,x1).fit()
results.summary()
```

```python
slope=0.7
intercept=-0.1

def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc,x))

print(mymodel)
```

```python
plt.scatter(x,y)
fig=plt.plot(x,mymodel,lw=3,c='purple',linestyle='-')
plt.xlabel("Advertising Cost")
plt.ylabel("Sales")
plt.show()
```

```python
#Fertilizer and Yield using OLS Model

x=[4,6,10,12]
y=[3,5.5,6.5,9]

plt.scatter(x,y)
plt.xlabel("Fertilizer")
plt.ylabel("Yield")
plt.show()
```

```python
x1=sm.add_constant(x)
results=sm.OLS(y,x1).fit()
results.summary()
```

```python
slope=0.65
intercept=0.8

def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc,x))

print(mymodel)
```

```python
plt.scatter(x,y)
fig=plt.plot(x,mymodel,lw=3,c='purple',linestyle='--')
plt.xlabel("Fertilizer")
plt.ylabel("Yield")
plt.show()
```
